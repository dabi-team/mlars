```python
import matplotlib.pyplot as plt
def plot_graph(accuracy, precision, recall, f1, average, x_labels,title):
    average_arrays = [accuracy, precision, recall, f1]
    print(average)
    print(average[9])
    highest_average = max(average)
    for index,i in enumerate(average):
        if i == highest_average:
            highest_average_index =index 

    print(highest_average)
    array = [highest_average for i in range(len(average))]
    
    plt.figure(figsize=(10,5))
    plt.xticks(rotation=90)
    plt.plot(x_labels, array,  label='Best Model (' + x_labels[highest_average_index] + ')', color='blue',  linestyle='dashed')
    plt.plot(x_labels, average,label="Average")
    plt.plot(x_labels, accuracy,label="Accuracy")
    plt.plot(x_labels, recall,label="Recall")
    plt.plot(x_labels, precision,label="Precision")
    plt.plot(x_labels, f1,label="F1")
    plt.title(title)
    # plt.yticks(rotation=90)
    plt.legend()
    plt.savefig("./finals/" +title + '.png')
    plt.show()
    # create final folder and save the plot
    # plt.savefig('./plots/' + title + '.png')
    plt.close()
```
