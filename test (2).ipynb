```python
import pandas as pd
import numpy as np
from mlars import main

df = pd.read_csv("datasets/heart.csv", encoding= 'unicode_escape')
resultdf = main(df, "HeartDisease") 

import matplotlib.pyplot as plt
```

```python
# plt.plot(resultdf["Name"], resultdf["model"],label="Best model (Random Forest)",linestyle='dashed')
print(len(resultdf["Name"]))
plt.xticks(rotation=90)
plt.plot(resultdf["Name"], resultdf["Average"],label="Average")
plt.plot(resultdf["Name"], resultdf["Accuracy"],label="Accuracy")
plt.plot(resultdf["Name"], resultdf["Recall"],label="Recall")
plt.plot(resultdf["Name"], resultdf["Precision"],label="Precision")
plt.plot(resultdf["Name"], resultdf["F1"],label="F1")
plt.title("Heart Disease")
plt.figure(figsize=(10,5))
# plt.yticks(rotation=90)

plt.legend()
plt.show()
```
